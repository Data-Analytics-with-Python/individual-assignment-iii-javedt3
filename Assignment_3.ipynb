# Question 1 (45 points)

In this question, we will predict a customer's spending amount on each product category over a two year period. Let us assume that when we try to predict a customer's spending on a product category (such as wines), their spending on other products is not observable.

In this question and Question 2, we will focus on **wines**.

(i). **Split the data** into two data frames, X (**features**) and y (**target**).

Then, further **split the data** into **training** and **testing** sets.

````python

y = df['MntWines']
X = df.drop(columns=['MntWines', 'ID', 'Dt_Customer'])

display(X.head())
display(y.head())
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=20)
````

(ii). **Build a machine learning pipeline** that combines the following steps to predict spending amount on wines:

* Performing one-hot encoding for the categorical features;
* A random forest model for regression.

In the random forest model, **specify** the following hyperparameters:
* Number of trees;
* Maximum depth of any tree
* Minimum number of data points required to split a node;
* Minimum number of data points in any leaf node

In addition, **fit your model** to the training data.

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline

categorical_features = ['Education', 'Marital_Status']
numerical_features = [col for col in X_train.columns if col not in categorical_features]

ct = ColumnTransformer(transformers=[("cat", OneHotEncoder(drop="first", sparse_output=False, handle_unknown='ignore'), categorical_features)
    ],remainder="passthrough")

rf = RandomForestRegressor(
    n_estimators=100,
    max_depth=10,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=20
)

pipe = Pipeline(
    steps=[
        ("preprocess", ct),
        ("model", rf)
    ]
)


pipe.fit(X_train, y_train)

pipe
````

(iv). **Consider** **two** measures to evaluate the model's performance on the test dataset.

````python
from sklearn.metrics import mean_absolute_error, r2_score

y_pred = pipe.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R2): {r2:.2f}")
````

Using Mean Absolute Error (MAE) and R² to evaluate the model, the results indicate strong predictive performance on the test data. The model achieves an MAE of 89.67, which means that the predicted wine-spending amounts differ from the actual values by about 90 dollars on average, indicating reasonably accurate individual predictions. The R² value of 0.78 shows that the model explains 78% of the variance in wine spending, suggesting that the model captures the main patterns in the data well.

Overall, the model performs well both in terms of prediction accuracy and goodness-of-fit, although some prediction error remains—particularly for extreme values. Nevertheless, the results indicate that the model is reliable and effective for predicting customers’ wine spending.

  * Note that the absolute residual is distance between the predicted value and actual value, i.e., $|y_{pred}-y_{test}|$.

So we end up with two sets of absolute residuals (one by the linear regression model and the other by the random forest model).

**Step 2**. For each pair of absolute residuals for the same test data point, we can define a point in a scatterplot. Genereate such a scatterplot using `plotly.express` (LR residuals vs. RF residuals).  

**Step 3**. Add a 45 degree reference line to the plot. This can be done using the following codes. (You may need to change `min_val` and `max_val` for better visualization).


**Implement the above steps**.

Note that the above steps essentially creates a [Q-Q plot](https://en.wikipedia.org/wiki/Q%E2%80%93Q_plot). How would you interpret the plot (for comparing the two predictive models).

````python
import numpy as np
import pandas as pd

abs_resid_lr = np.abs(y_pred_lr - y_test)
abs_resid_rf = np.abs(y_pred_rf - y_test)

resid_df = pd.DataFrame({
    "LR_abs_resid": abs_resid_lr,
    "RF_abs_resid": abs_resid_rf
})

display(resid_df.head())
````

````python
import plotly.express as px

fig = px.scatter(
    resid_df,
    x="LR_abs_resid",
    y="RF_abs_resid",
    labels={
        "LR_abs_resid": "Linear Regression |Residual|",
        "RF_abs_resid": "Random Forest |Residual|"
    },
    title="Absolute Residuals: Linear Regression vs Random Forest"
)

fig.show()
````

````python
min_val = 0
max_val = max(resid_df["LR_abs_resid"].max(),
              resid_df["RF_abs_resid"].max()) * 1.05   # slight padding

fig.add_shape(
    type="line",
    x0=min_val, y0=min_val,
    x1=max_val, y1=max_val,
    line=dict(color="red", width=2, dash="dash"),
)

fig.update_xaxes(range=[min_val, max_val])
fig.update_yaxes(range=[min_val, max_val])

fig.show()
````

This plot compares the absolute prediction errors of the linear regression model (x-axis) with those of the random forest model (y-axis) for each customer in the test set. The dashed red 45-degree line represents cases where both models make the same size error on a given observation. Points below the line indicate that the random forest is more accurate than the linear regression for that customer. Points above the line indicate that linear regression performs better for that customer. Points on the line represent equal performance by both models. In this plot, the majority of points lie below the reference line, showing that the random forest generally has smaller absolute residuals than linear regression for most customers. This confirms that the random forest model provides more accurate predictions overall. Additionally, for observations where linear regression makes large errors (far to the right on the x-axis), random forest often shows much smaller corresponding errors (lower y-values). This indicates that random forest is not only better on average, but also more accurate for extreme cases, where linear regression tends to perform poorly. A small number of points appear above the line, meaning there are cases where linear regression performs better, but these are relatively limited and tend to be small-error cases.

# Question 3. (24 points)

In this question, we will consider a classification problem on customers' spendings on meat products.


(i). For the column that represents customers' spendings on meat products, **calculate** the 33.33% and 66.67% percentiles. (**Hint**: You may use the function `df['MntMeatProducts'].quantile([1/3,2/3])`.)

Based on the two percentiles, **label** each row in the dataset
* If a customer's spending is below the 33.33% percentile, label their spending as "low";

* If a customer's spending is above the 66.67% percentile, label their spending as "high";


* If a customer's spending is between the two percentiles, label their spending as "medium".

````python
percentiles = df['MntMeatProducts'].quantile([1/3, 2/3])

p33 = percentiles.loc[1/3]
p67 = percentiles.loc[2/3]

print("33.33% percentile:", p33)
print("66.67% percentile:", p67)
````

````python
def label_meat_spending(x):
    if x < p33:
        return "low"
    elif x > p67:
        return "high"
    else:
        return "medium"

df['MeatSpendingLabel'] = df['MntMeatProducts'].apply(label_meat_spending)

df['MeatSpendingLabel'].value_counts()
````

````python
df[['MntMeatProducts', 'MeatSpendingLabel']].head(10)
````

\(ii). **Build a K-nearest-neighbors model** to predict the label of a customer's spending on meat products. The model should be part of a machine learning pipeline that preprocesses the data.





````python
from sklearn.neighbors import KNeighborsClassifier

y_meat = df['MeatSpendingLabel']
X_meat = df.drop(columns=['MntMeatProducts', 'ID', 'Dt_Customer', 'MeatSpendingLabel'])

X_meat_train, X_meat_test, y_meat_train, y_meat_test = train_test_split(X_meat, y_meat, test_size=0.2, random_state=20, stratify=y_meat)

categorical_features_meat = ['Education', 'Marital_Status']
numerical_features_meat = [col for col in X_meat_train.columns if col not in categorical_features_meat]

ct_meat = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(sparse_output=False, drop="first", handle_unknown='ignore'), categorical_features_meat),
    ],
    remainder="passthrough")

knn = KNeighborsClassifier()

pipeline_knn = Pipeline(steps=[
    ('preprocess', ct_meat),
    ('classifier', knn)
])

pipeline_knn.fit(X_meat_train, y_meat_train)

````

(iii). Evaluate the performace of your model in part (ii) on the test data by **generating the classification report**.

Further, **interpret** each number in the classification report based on the current context.

````python
from sklearn.metrics import classification_report

y_meat_pred = pipeline_knn.predict(X_meat_test)

report = classification_report(y_meat_test, y_meat_pred)

print(report)
````
